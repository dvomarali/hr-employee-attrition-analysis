```python
plt.figure(figsize=(12,8))

sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
df["AgeGroup"] = pd.cut(df["Age"], bins=[18,25,35,45,60], labels=["18-25","26-35","36-45","46-60"])

sns.countplot(x="AgeGroup", hue="Attrition", data=df)

plt.title("Attrition by Age Group")
plt.show()
```

# Insight
Employees aged 26–35 show higher attrition compared to older employees.

```python
sns.scatterplot(x="TotalWorkingYears", y="MonthlyIncome", hue="Attrition", data=df)

plt.title("Income vs Experience with Attrition")
plt.show()
```

# Insight:

Employees with lower experience and lower income appear more likely to leave the company.

```python
attrition_rate = df.groupby("Department")["Attrition"].value_counts(normalize=True).unstack()

attrition_rate["Yes"].plot(kind="bar")

plt.title("Attrition Rate by Department")
plt.ylabel("Attrition Rate")
plt.show()
```

# Insight:

The Sales department shows the highest attrition rate compared to other departments.

# Final Insights
Employees working overtime are more likely to leave the company.

Younger employees show higher attrition rates.

Lower income levels are associated with higher employee turnover.

Some departments experience higher attrition compared to others.
